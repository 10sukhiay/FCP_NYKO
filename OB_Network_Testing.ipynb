## Initial testing of networks and how they could be applied to the project

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python
# creating a simple graph
G=nx.Graph()
# add_nodes_from takes a list as input and creates new nodes based 
# off the list
G.add_nodes_from(['A','B', 'C'])
# add_edge creates an edge between the nodes
G.add_edge('A','B')
G.add_edge('A','C')

# give attributes to nodes and edges
G.nodes['A']['Number'] = 12
G.nodes['B']['Number'] = 14
G.nodes['C']['Number'] = 8
G.edges['A','B']['Time'] = 5

nx.draw(G, with_labels=True)
```

### Querying data

```python
# Return list of nodes 
G.nodes()
```

```python
# to see nodes and attributes - dictionary
G.nodes(data=True)
```

```python
# Retuns a list of tuples. Each tuple represents an edge
G.edges()
```

```python
# Return edges and attributes - dictionary
G.edges(data=True)
```
